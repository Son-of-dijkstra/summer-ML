```python
# RGB images 32x32
model = nn.Sequential(
    nn.Conv2d(3, 6, 3, padding = 1),
    nn.ReLU(),
    nn.Conv2d(6, 6, 3, padding = 1),
    nn.ReLU(),
    nn.MaxPool2d(2), # 6x16x16
    nn.Conv2d(6, 12, 3, padding = 1),
    nn.ReLU(),
    nn.Conv2d(12, 12, 3, padding = 1),
    nn.Flatten(0, -1), #12*16x16
    nn.Linear(12*16*16, 128),
    nn.ReLU(),
    nn.Linear(128, 10), # 10 classes
)
```

```python
params = list(model.parameters())
```

```python
sumparam = sum(p.numel() for p in params)
sumparam
```

```python
img = torch.randn((3, 32, 32))
```

```python
model(img)
```
