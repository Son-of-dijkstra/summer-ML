```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from micrograd.autograd_engine import Value
from micrograd.nn_module import Neuron, Layer, MLP
```

```python
np.random.seed(42)
random.seed(42)
```

```python
from sklearn.datasets import load_iris

data = load_iris()
target = data['target']
mask = target!=2
target = target[mask]
X = data['data'][mask]
```

```python
model = MLP(4, (8, 4, 1), (None, 'relu', 'sigmoid'))
model
```

```python
def getloss(probs):
    return sum((prob-gt)**2 for prob, gt in zip(probs, target))/len(probs)
```

```python
epochs = 200
lr = 0.05
for epoch in range(epochs + 1):
    probs = [model(X[i]) for i in range(len(target))]
    loss = getloss(probs)
    if epoch%10==0:
        print(f'Epoch: {epoch+1}, loss: {loss.data}')

    model.zero_grad()

    loss.backward()

    for p in model.parameters():
        p.data += -lr * p.grad
```
